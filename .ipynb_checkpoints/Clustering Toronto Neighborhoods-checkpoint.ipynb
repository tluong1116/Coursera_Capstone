# Clustering Toronto Neighborhoods

In this project, we will explore, segment, and cluster the neighborhoods in the city of Toronto.

For the Toronto neighborhood data, a Wikipedia page exists that has all the information we need to explore and cluster the neighborhoods in Toronto. We will scrape the Wikipedia page using BeautifulSoup and wrangle the data, clean it using Pandas.

Finally, we will segmentize the neighborhoods using the venues data provided by Foursquare, via its API.

## Parse the wikipedia page with BeautifulSoup

### Getting the data

```python
# Get content of the page
page_content = requests.get('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M').content
# Get
page_parsed = BeautifulSoup(page_content, 'html.parser')
```

```python
# Print out prety html page
print(page_parsed.prettify())
```

Since there is only one table on the page, we can find the table using the tag 'table'. Reading the html we also see that its class is the 'wikitable' class.

```python
table = page_parsed.find_all('table', class_ ="wikitable")[0]
```

Since all the rows have tag 'tr' accompanied them, we can find them using tag 'tr', then loop over them and using tag 'td' to parse the data into a dataframe.

```python
table_rows = table.find_all('tr')
```

```python
# Create empty list to hold the cells of the table
postal_code = []
borough = []
hood = []

for i in range(1, len(table_rows)):
    postal_code.append(table_rows[i].find_all('td')[0].get_text().strip())
    borough.append(table_rows[i].find_all('td')[1].get_text().strip())
    hood.append(table_rows[i].find_all('td')[2].get_text().strip())

# Create dictionary to make data frame
toronto_hood_dict = {
    'postal_code':postal_code,
    'borough':borough,
    'neighborhood':hood
}

toronto_hood_df = pd.DataFrame(toronto_hood_dict)
print(toronto_hood_df['borough'].value_counts())
toronto_hood_df.head()
```

Since want both the Borough and the Neighborhood in our data, we will filter out the 'Not assigned' rows from the data frame.

```python
toronto_df = toronto_hood_df[toronto_hood_df['borough'] != 'Not assigned'].reset_index(drop=True)
print(toronto_df['borough'].value_counts())
toronto_df.head()
```

```python
# Change 'Not assigned' neighborhood to be the same as its borough
toronto_df['neighborhood'][toronto_df['neighborhood'] == 'Not assigned'] = toronto_df['borough']
```

```python
LIMIT = 100 # limit of number of venues returned by Foursquare API
radius = 500 # define radius
def getNearbyVenues(names, latitudes, longitudes, radius=500):
    
    venues_list=[]
    for name, lat, lng in zip(names, latitudes, longitudes):
        print(name)
            
        # create the API request URL
        url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            VERSION, 
            lat, 
            lng, 
            radius, 
            LIMIT)
            
        # make the GET request
        results = requests.get(url).json()["response"]['groups'][0]['items']
        
        # return only relevant information for each nearby venue
        venues_list.append([(
            name, 
            lat, 
            lng, 
            v['venue']['name'], 
            v['venue']['location']['lat'], 
            v['venue']['location']['lng'],  
            v['venue']['categories'][0]['name']) for v in results])

    nearby_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
    nearby_venues.columns = ['Neighborhood', 
                  'Neighborhood Latitude', 
                  'Neighborhood Longitude', 
                  'Venue', 
                  'Venue Latitude', 
                  'Venue Longitude', 
                  'Venue Category']
    
    return(nearby_venues)
```

```python
def return_most_common_venues(row, num_top_venues):
    row_categories = row.iloc[1:]
    row_categories_sorted = row_categories.sort_values(ascending=False)
    
    return row_categories_sorted.index.values[0:num_top_venues]
```

#### Determine best k value

To determine the best k for our KMeans algorithm, we will use the elbow method. We define the following:

- **Distortion**: It is calculated as the average of the squared distances from the cluster centers of the respective clusters. Typically, the Euclidean distance metric is used. 
- **Inertia**: It is the sum of squared distances of samples to their closest cluster center.

We will iterate the value of k from 2 to 10 and calculate the values of distortions for each value of k and calculate the distortion and inertia for each value of k in the given range.

```python
distortions = [] 
inertias = [] 
mapping1 = {} 
mapping2 = {} 
k_values = range(2,20)

toronto_grouped_clustering = toronto_grouped.drop('Neighborhood',axis=1)

for k in k_values: 
    #Building and fitting the model 
    kmeanModel = KMeans(n_clusters=k, random_state=0).fit(toronto_grouped_clustering) 

    distortions.append(sum(np.min(cdist(toronto_grouped_clustering, kmeanModel.cluster_centers_, 
                      'euclidean'),axis=1)) / toronto_grouped_clustering.shape[0]) 
    inertias.append(kmeanModel.inertia_) 
  
    mapping1[k] = sum(np.min(cdist(toronto_grouped_clustering, kmeanModel.cluster_centers_, 
                 'euclidean'),axis=1)) / toronto_grouped_clustering.shape[0] 
    mapping2[k] = kmeanModel.inertia_ 
    

# Print out distortion
print('Distortion:')
for key,val in mapping1.items(): 
    print(str(key)+' : '+str(val)) 

print("===================")

# Print out inertial
print('Inertial')
for key,val in mapping2.items(): 
    print(str(key)+' : '+str(val)) 
```

```python
# Plot Distortion and Inertial to find the optimal k
fig,(ax1,ax2) = plt.subplots(1, 2, figsize = (16,4))

ax1.plot(k_values, distortions, 'bx-') 
ax1.set_xlabel('Values of K') 
ax1.set_xticks(list(range(2,20)))
ax1.set_ylabel('Distortion')
ax1.grid(axis='x')
ax1.set_title('The Elbow Method using Distortion')

ax2.plot(k_values, inertias, 'rx-') 
ax2.set_xlabel('Values of K') 
ax2.set_xticks(list(range(2,20)))
ax2.set_ylabel('Inertia') 
ax2.grid(axis='x')
ax2.set_title('The Elbow Method using Inertia') 

plt.show() 
```

While the elbow shape is not clear, we can see that the slop of the Distortion is the steepest at k = 6. We will pick 6 as our k value for clustering.

#### Clustering with k = 6

```python
# set number of clusters
kclusters = 6

# run k-means clustering
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(toronto_grouped_clustering)

# check cluster labels generated for each row in the dataframe
kmeans.labels_[0:10] 
```

### Observations

We can see that majority of the neighborhoods are in cluster 0, with only 1 of neighborhood in other clusters. Cluster 0 contains the urban area, with a lot of shops. The other clusters look like suburban or residential areas.

The concentrated of neighborhood in  might be because our segmentation criteria on all venues are too wide, and we are suffering from ('Cusse of Dimensionality')[https://en.wikipedia.org/wiki/Curse_of_dimensionality]. When there are too many features in the dataset, it's very hard to perform clustering right since the distance between all of the points appear to be very similar. We can also see from our Distortion and Inertia plot.

Since the assignment end here, we will stop. However, to overcome this problem in the future, we can try out approaches like:

- Narrow down the segmentation criteria. This might include perform clustering on cluster 0 one more time.
- Perform dimensionality reduction algorithm
